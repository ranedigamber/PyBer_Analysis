```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of string for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create a plot
plt.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
#Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="o", color="blue", linewidth=2, label="Boston")

#Create lables for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()

#Add gridlines
plt.grid()
```

```python
#Creating scatter plots


# Set the x-axis to a list of string for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]


plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(y_axis, x_axis, color = "red", label="Chicago")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Create a new list, a for loop to iterate throught y_axis list, append the new list so that each items in the original series is 
#multipled by 3
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Refactoring the above cell using Python list comprehension
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
#Skill drill
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "red", s=[i*5 for i in y_axis], label = "Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.legend()
```

```python
#Creating pie chart
#fig, ax = plt.subplots()
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, labels=x_axis, autopct='%.1f%%', explode=explode_values)
plt.show()
```
